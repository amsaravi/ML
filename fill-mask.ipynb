```python
df = pd.read_csv('bbc_text_cls.csv')
```

```python
labels = set(df['labels'])
```

```python
texts = df[df['labels']=='business']['text']
```

```python
texts
```

```python
np.random.seed(1234)
```

```python
texts.shape
```

```python
i = np.random.choice(texts.shape[0])
```

```python
doc = texts.iloc[i]
```

```python
doc
```

```python
wp_text = textwrap.fill(doc,replace_whitespace=False,fix_sentence_endings=True)
print(wp_text)
```
